# Gradient Descent Code

```python
def mse_loss_fxn(x, w, bias, y):
  yhat = x@w+bias
  if yhat.shape != y.shape:
    print('shape mismatch!')
  else:
    mse = np.sum((x @ w + bias - y)**2)/y.shape[0]
  return mse
```

```python
def bias_derivative(x, w, bias, y):
  myderivative = np.sum(x @ w + bias - y)/y.shape[0] # x is a vector, w is a scalar and y is a vector
  return myderivative
```

```python
def dependent_var_derivative(x, w, bias, y):
  if x.shape[1] == 1:
    if len(w.shape) != 2:
      w = w.reshape(-1,1)
    myderivative = np.sum(x.T @ (x @ w + bias - y))/y.shape[0] # x is a vector, w is a scalar and y is a vector
    return myderivative
  else:
    print('x should be a vector of shape m,1 but found',x.shape)
```

```python
def gradient_descent_update_dep_var(x, w, bias, y, alpha):
  if len(w.shape) != 2:
      w = w.reshape(-1,1)
  w = w - alpha * (dependent_var_derivative(x, w, bias, y))
  return w
def gradient_descent_update_bias(x, w, bias, y, alpha):
  bias = bias - alpha * (bias_derivative(x, w, bias, y))
  return bias
```

## Participation block

what do you see little change in our graph?

**Ans:** The minimal change in the graph likely results from a small learning rate or gradients that approach zero, causing very slow updates. As a result, the model’s predictions and error reduction occur so incrementally that they barely shift the line or alter the MSE, making progress appear stagnant.
<br>

why is w going where we want but not our bias?

**Ans:** The weight w progresses toward the optimal value because it directly adjusts in response to the input-output relationship, but bias updates depend on the sum of residuals. When residuals are close to zero, the bias gradient becomes small, resulting in minimal or no updates to the bias and limiting its contribution to model improvement.
